```python
import gpt_2_simple as gpt2
from datetime import datetime
```

```python
json_file = 'realdonaldtrump.json'
text_file = 'realdonaldtrump.txt'

import json
import re

def strip_non_ascii(string):
    ''' Returns the string without non ASCII characters'''
    stripped = (c for c in string if 0 < ord(c) < 127)
    return ''.join(stripped)

with open(json_file, encoding='utf8') as data_file:    
    data = json.load(data_file)
    outputfile = open(text_file,"w+", encoding='utf8')
    for tweet in data:
    #if '2010' in tweet['created_at']:
        #print(tweet)
        #print(tweet['full_text'])
        #start = '<|startoftext|>'
        full_text = tweet['full_text']
        full_text = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', full_text)
        if '2012' in tweet['full_text']:
            full_text.replace('2012', '2020')
        if '2016' in tweet['full_text']:
            full_text.replace('2016', '2020')

        #end = '<|endoftext|>'
        #full_text = start + full_text + end
        print(full_text)

        outputfile.write(str(full_text))
    outputfile.close() 
    
```

```python
sess = gpt2.start_tf_sess()
```

```python
!ln -s ../models .
```

```python
file_name = text_file
gpt2.finetune(sess,
            dataset=file_name,
            model_name='355M',
            steps=10,
            restore_from='fresh',
            run_name='run1',
            print_every=1,
            sample_every=1,
            save_every=5)
```

```python
prefix = "@realDonaldTrump Bring back Celebrity Apprentice!"
output = gpt2.generate(sess,
             length=240,
             temperature=0.95,
             prefix=prefix,
             nsamples=20,
             batch_size=5,
             run_name='run1',
             truncate="<|endoftext|>",
             include_prefix=False)

output
```
