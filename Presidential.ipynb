```python
import gpt_2_simple as gpt2
from datetime import datetime
```

```python
json_file = 'realdonaldtrump.json'
text_file = 'realdonaldtrump.txt'

import json

def strip_non_ascii(string):
    ''' Returns the string without non ASCII characters'''
    stripped = (c for c in string if 0 < ord(c) < 127)
    return ''.join(stripped)

with open(json_file, encoding='utf8') as data_file:    
    data = json.load(data_file)
    outputfile = open(text_file,"w+", encoding='utf8')
    for tweet in data:
        #print(tweet['full_text'])
        start = '<|startoftext|>'
        full_text = str(strip_non_ascii(tweet['full_text']))
        end = '<|endoftext|>'
        full_text = start + full_text + end
        print(full_text)
        
        outputfile.write(str(full_text))
    outputfile.close() 
    
```

```python
sess = gpt2.start_tf_sess()
```

```python
!ln -s ../models .
```

```python
file_name = text_file
gpt2.finetune(sess,
            dataset=file_name,
            model_name='355M',
            steps=10,
            restore_from='fresh',
            run_name='run1',
            print_every=1,
            sample_every=1,
            save_every=5)
```

```python
prefix = "\
Today is Jan 3, 2020. Donald Trump is president. He is confronted by a child on the East Lawn. \
\
\
CHILD: My name is Parker, I'm in the 2nd grade, what would you do if you were younger? \
\
DONALD TRUMP: Thank you for the question. "
output = gpt2.generate(sess,
             length=240,
             temperature=0.95,
             prefix='<|startoftext|>'+prefix,
             nsamples=20,
             batch_size=20,
             run_name='run1',
             truncate="<|endoftext|>",
             include_prefix=False)

output
```
