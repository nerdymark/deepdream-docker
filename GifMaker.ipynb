```python
import os, sys, re

import imageio

filenames = sorted([f for f in os.listdir('.') if re.match(r'[0-9]+.*\.jpg', f)])
print(filenames)

kargs = { 'fps': 60 }

images = []
for filename in filenames:
    images.append(imageio.imread(filename))
imageio.mimsave('movie.gif', images, **kargs)
```
