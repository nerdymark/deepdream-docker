```python
import json
import re
from os import walk
from nltk.tokenize import word_tokenize


def Convert(string): 
    li = list(string.split(" "))
    return li

topic = "mueller"

with open("realdonaldtrump.json", "r") as tweetfile:
    tweets = json.load(tweetfile)
    wordlist = []
    for tweet in tweets:
        words = Convert(tweet['full_text'])
        if topic in tweet['full_text'].casefold() and tweet['user']['screen_name'].casefold() == 'realdonaldtrump':
            print(tweet['full_text'])
            for word in words:
                word = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", word)
                word = re.sub('&amp;', '&', word)
                word = re.sub('&quot;', '"', word)
                word = re.sub('&apos;', "'", word)
                word = re.sub('&gt;', '>', word)
                word = re.sub('&lt;', '<', word)
                word = re.sub('"', '', word)
                word = re.sub('"', '', word)
                word = re.sub(r"'", '', word)
                word = re.sub('"', '', word)
                word = re.sub(r'\\n', '', word)
                word = re.sub(r':', '', word)
                word = re.sub('@', '', word)
                word = re.sub('-', '', word)
                word = re.sub('\n', '', word)
                word = re.sub('…', '', word)
                word = re.sub(r'\(', '', word)
                word = re.sub(r'\)', '', word)
                word = re.sub('\.', '', word)
                word = re.sub('&', '', word)
                word = re.sub('\,', '', word)
                word = re.sub('\?', '', word)
                word = re.sub('\!', '', word)
                word = re.sub("\'", '', word)
                word = re.sub("”", '', word)
                word = re.sub("“", '', word)
                word = re.sub("’", '', word)
                word = re.sub("'", '', word)
                word = re.sub("•", '', word)
                word = re.sub("%", '', word)
                word = re.sub("`", '', word)
                if len(word) > 2:
                    wordlist.append(word)



```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt 
import time
import random
from os import walk, path
import os
import numpy as np
from PIL import Image


d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

#outFile = 'wordclouds/'+topic.upper()+'.png'
outFile = 'wordclouds/trump_mueller.png'
origMask = 'masks/mueller1.jpg'

resizedFile = '/tmp/tmpResize.jpg'

basewidth = 1024
img = Image.open(origMask)
wpercent = (basewidth/float(img.size[0]))
hsize = int((float(img.size[1])*float(wpercent)))
img = img.resize((basewidth,hsize), Image.ANTIALIAS)
img.save(resizedFile)

image_coloring = np.array(Image.open(path.join(d, resizedFile)))



fontpath = 'fonts/'
fontlist = []
for (dirpath, dirnames, filenames) in walk(fontpath):
    fontlist.extend(filenames)
    break
font = random.choice(fontlist)
font = fontpath+font


comment_words = ' '
stopwords = set(STOPWORDS)
  
# iterate through the json file 

for val in df[0]: 
      
    # typecaste each val to string 
    val = str(val)
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
  
  
print("Word count:", len(comment_words))

wc = WordCloud(width = 1024, height = 1024, 
                background_color ='white', 
                stopwords = stopwords,
                max_words = 6765,
                font_path = font,
                mask = image_coloring,
                min_font_size = 3,
                random_state = 42).generate(comment_words)

image_colors = ImageColorGenerator(image_coloring)




  
# show
fig, axes = plt.subplots(1, 3)
axes[0].imshow(wc, interpolation="bilinear")
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
axes[1].imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
axes[2].imshow(image_coloring, cmap=plt.cm.gray, interpolation="bilinear")
for ax in axes:
    ax.set_axis_off()
plt.show()
  
wc.to_file(outFile)
```
